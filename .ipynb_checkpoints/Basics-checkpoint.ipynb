## Hypothesis testing about mean difference

Null hypothesis: mean of fandango and metacritic norm should be close each other
Alternative hypothesis: there is something suspicious about fandango ratings

Our test statistics is mean difference.

```python
mean_difference = movies["Fandango_Stars"].mean() - movies["Metacritic_norm_round"].mean()
```

```python
mean_difference
```

```python
all_values = pd.concat([ movies["Fandango_Stars"],movies["Metacritic_norm_round"]])

mean_differences = []

for i in range(0,1000):
    group_a = []
    group_b = []
    for value in all_values:
        assingment_chance = np.random.rand()
        if assingment_chance >= 0.5:
            group_a.append(value)
        else:
            group_b.append(value)
    iter_mean_diff = np.mean(group_b) - np.mean(group_a)
    mean_differences.append(iter_mean_diff)

plt.hist(mean_differences)
plt.show()
```

```python
sampling_distribution = {}
for df in mean_differences:
    if sampling_distribution.get(df, False):
        sampling_distribution[df] = sampling_distribution[df] + 1
    else:
        sampling_distribution[df] = 1
```

```python
frequencies = []

for sp in sampling_distribution.keys():
    if sp >= mean_difference:
        frequencies.append(sampling_distribution[sp])
p_value = np.sum(frequencies) / 1000
```

```python
p_value
```

Since the p value of 0 is less than the threshold we set of 0.05, we conclude that the difference in mean ratings can't be attributed to random chance alone. We therefore reject the null hypothesis and accept the alternative hypothesis.
